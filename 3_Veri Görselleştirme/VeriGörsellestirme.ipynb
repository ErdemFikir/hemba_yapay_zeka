```python
#Kütüphaneleri içeri dahil ediyoruz
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
#Veriyi içeri aktaralım
veri = pd.read_csv("olimpiyatlar_temizlenmis.csv")
veri.head()
```

```python
#Saçılım Grafiği 
def sacilimGrafik():
    sns.scatterplot(x='boy', y='kilo', data=veri)
    plt.title("Boy ve Kilo Dağılım - Beyaz Izgara Tema")
    plt.show()
```

```python
sacilimGrafik()
```

```python
# Stili belirleme
sns.set_style("white")
sacilimGrafik()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")

# Plotly Kütüphanesi
from plotly.offline import init_notebook_mode, iplot, plot
import plotly as py 
init_notebook_mode(connected=True)
import plotly.graph_objects as go #graphical objects
import plotly.io as pio 
import plotly.figure_factory as ff 
```

```python
#İlk 100 Üniversite üniversiteyi filtreleyelim
df_100Uni = df[:100]
```

```python
#ilk Grafik  - Çizgi Grafik

cizgi1 = go.Scatter(
                    x= df_100Uni.dunya_siralamasi, # x ekseninde dünya sıralaması olsun
                    y=df_100Uni.alinti, # y eksenin alıntı olsun 
                    mode = "lines+markers", #grafik çizgi ve saçılım olsun
                    name ="Alıntı",  # etiket alıntı olsun
                    marker = dict(color='rgba(78,78,250,0.85)'), #renk seçimi yapıldı
                    text = df_100Uni.uni_isim #imleç ilgili noktaya geldiğinde üni adı yazılsın
                    )

cizgi2 = go.Scatter(
                    x= df_100Uni.dunya_siralamasi, # x ekseninde dünya sıralaması olsun
                    y=df_100Uni.ogretim, # y eksenin öğretim olsun 
                    mode = "lines+markers",  #grafik çizgi ve saçılım olsun
                    name ="Öğretim",  #etiket öğretim olsun
                    marker = dict(color='rgba(202,43,15,0.85)'), #renk seçimi yapıldı
                    text = df_100Uni.uni_isim  #imleç ilgili noktaya geldiğinde üni adı yazılsın
                    ) 

veri_ = [cizgi1,cizgi2] # 2 grafiği veri_ değişkeni içerisine koy

#başlık ekle ve eksen ayarlarını yap
yerlesim = dict(    title= 'Dünya Sıralamasında İlk 100 Üniversitenin Alıntı ve Öğretim Puanları',
                    xaxis= dict(title= 'Dünya Sıralaması', 
                                ticklen=5, 
                                zeroline=False)
                )

fig = dict(data = veri_, layout=yerlesim) #Figür oluştur             
plot(fig,filename = '1_çizgi-alıntı ve öğretim puanları.html') #.html olarak kaydet
```

```python
#Saçılım Grafiği
#2014,2015,2016 yıllarını ilk 100 üniversite için ayıralım

veri2014 = df[df.yil == 2014].iloc[:100,:]
veri2015 = df[df.yil == 2015].iloc[:100,:]
veri2016 = df[df.yil == 2016].iloc[:100,:]

#İlk Grafik
cizgi1 = go.Scatter(
                    x= veri2014.dunya_siralamasi, # x ekseninde dünya sıralaması olsun
                    y=veri2014.alinti, # y eksenin alıntı olsun 
                    mode = "markers", #grafik çizgi ve saçılım olsun
                    name ="2014",  # etiket alıntı olsun
                    marker = dict(color='rgba(2,255,15,0.8)'), #renk seçimi yapıldı
                    text = veri2014.uni_isim #imleç ilgili noktaya geldiğinde üni adı yazılsın
                    )

#İkinci Grafik
cizgi2 = go.Scatter(
                    x= veri2015.dunya_siralamasi, # x ekseninde dünya sıralaması olsun
                    y=veri2015.alinti, # y eksenin alıntı olsun 
                    mode = "markers", #grafik çizgi ve saçılım olsun
                    name ="2015",  # etiket alıntı olsun
                    marker = dict(color='rgba(2,2,255,0.8)'), #renk seçimi yapıldı
                    text = veri2015.uni_isim #imleç ilgili noktaya geldiğinde üni adı yazılsın
                    )

#Üçüncü Grafik
cizgi3 = go.Scatter(
                    x= veri2016.dunya_siralamasi, # x ekseninde dünya sıralaması olsun
                    y=veri2016.alinti, # y eksenin alıntı olsun 
                    mode = "markers", #grafik çizgi ve saçılım olsun
                    name ="2016",  # etiket alıntı olsun
                    marker = dict(color='rgba(255,2,15,0.8)'), #renk seçimi yapıldı
                    text = veri2016.uni_isim #imleç ilgili noktaya geldiğinde üni adı yazılsın
                    )


veri_ = [cizgi1,cizgi2,cizgi3] # 3 grafiği veri_ değişkeni içerisine koy

#başlık ekle ve eksen ayarlarını yap
yerlesim = dict(    title= '2014,2015,2016 Dünya Sıralamasında İlk 100 Üniversitenin Alıntı ve Öğretim Puanları',
                    xaxis= dict(title='Dünya Sıralaması', 
                                ticklen=5, 
                                zeroline=False),

                    yaxis= dict(title='Alıntı',
                                ticklen=5,
                                zeroline=False
                                )
                )

fig = dict(data = veri_, layout=yerlesim) #Figür oluştur             
plot(fig,filename = '2_sacılım-alıntı ve öğretim puanları.html') #.html olarak kaydet
```
